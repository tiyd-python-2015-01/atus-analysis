Question #1: Starting with something simple.... how much do unemployed vs employed people clean on average?

```python
unem = (base_data.TELFS >= 3)
em = (base_data.TELFS <= 2)
```

```python
unemployed = base_data[unem]
employed = base_data[em]
```

```python
unemployed_cleaning = unemployed[['TUFINLWGT','t020101']]
employed_cleaning = employed[['TUFINLWGT','t020101']]
```

```python
def activity_columns(data, activity_code):
    """For the activity code given, return all columns that fall under that activity."""
    col_prefix = "t{}".format(activity_code)
    return [column for column in data.columns if re.match(col_prefix, column)]
```

```python
def average_minutes2(data, activity_code):
    cols = activity_columns(data, activity_code)
    activity_data = data[cols]
    activity_sums = activity_data.sum(axis=1)
    data = data[['TUFINLWGT']]
    data['minutes'] = activity_sums
    data = data.rename(columns={"TUFINLWGT": "weight"})
    data['weighted_minutes'] = data.weight * data.minutes
    return data.weighted_minutes.sum() / data.weight.sum()
```

```python
u_cleaning = average_minutes2(unemployed_cleaning,'020101')
```

```python
e_cleaning = average_minutes2(employed_cleaning,'020101')
```

```python
cleaning = pd.DataFrame([u_cleaning, e_cleaning])
cleaning.index = ["Unemployed","Employed"]
cleaning.columns = ["Time Spent Cleaning"]
```

```python
cleaning.plot(kind = 'bar')
```

Do people who exercise more have higher general life satisfaction?

Importing the data files... 

```python
affect_data = pd.read_csv("atusdata/wbresp_2013/wbresp_2013.dat")
```

```python
affect_data.head()
```

```python
base_data.head()
```

Merging activity files with wellbeing.. 

```python
wb_merge = pd.merge(affect_data, base_data, left_on="TUCASEID", right_on="tucaseid")
```

```python
wb_merge.head()
```

First I need to select exercise variables and dichotimize them along the median.  

```python
wb_merge["Total Exercise"] = (wb_merge[activity_columns(wb_merge, "1301")].sum(axis = 1))
```

```python
wb_merge["Total Exercise"].median(), wb_merge["Total Exercise"].mean()
```

Apparently not many peple participated in the exercise activities listed on that day. I'll dichotomize at 0 (did or did not do physical activity).

```python
active = (wb_merge["Total Exercise"] >= 1)
did_ex = wb_merge[active]
didnt_ex = wb_merge[active == False]
```

```python
did_ex.info()
```

```python
didnt_ex.info()
```

Checking if they have similar variances on reported life satisfaction..

```python
did_ex["WECANTRIL"].var()
```

```python
didnt_ex["WECANTRIL"].var()
```

With unequal sample sizes: ( 1861, vs 8517), with similar variances: 4.27/3.07 < 2. 

Taking a look at the means.. 


```python
did_ex["WECANTRIL"].mean()
```

```python
didnt_ex["WECANTRIL"].mean()
```

Since the means are very similar I'll test to see if they are significantly different. 

```python
from scipy import stats as st
```

```python
st.ttest_ind(did_ex["WECANTRIL"],didnt_ex["WECANTRIL"],equal_var = True)
```

With p < .05, its safe to conclude that on average people who participated in physical activity the are more likely to report higher life satisfaction. 

```python
exercised = did_ex["WECANTRIL"].mean()
noexercise = didnt_ex["WECANTRIL"].mean()
Avg_satisfaction = pd.DataFrame([exercised, noexercise])
Avg_satisfaction.index = ["Exercised", "Didn't Exercise"]
Avg_satisfaction.columns = ["Average Reported Life Satisfaction"]
Avg_satisfaction.plot(kind = "bar",ylim = 6)

```

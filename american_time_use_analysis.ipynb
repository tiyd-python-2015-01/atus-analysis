# Function for Average Daily Minutes by Specific Category

```python
def average_minutes(data, activity_code):
    activity_col = "t{}".format(activity_code)
    data = data[['TUFINLWGT', activity_col]]
    data = data.rename(columns={"TUFINLWGT": "weight", activity_col: "minutes"})
    data['weighted_minutes'] = data.weight * data.minutes
    return data.weighted_minutes.sum() / data.weight.sum()
```

```python
sleeping = average_minutes(adults_with_no_children, "010101")
sleepless = average_minutes(adults_with_no_children, "010102")
(sleeping + sleepless) / 60
```

# Function for Average Daily Minutes by General Category

```python
def activity_columns(data, activity_code):
    """For the activity code given, return all columns that fall under that activity."""
    col_prefix = "t{}".format(activity_code)
    return [column for column in data.columns if re.match(col_prefix, column)]
```

```python
activity_columns(summary, "0101")
```

```python
def average_minutes2(data, activity_code):
    cols = activity_columns(data, activity_code)
    activity_data = data[cols]
    activity_sums = activity_data.sum(axis=1)
    data = data[['TUFINLWGT']]
    data['minutes'] = activity_sums
    data = data.rename(columns={"TUFINLWGT": "weight"})
    data['weighted_minutes'] = data.weight * data.minutes
    return data.weighted_minutes.sum() / data.weight.sum()
```

# Spending Time With Drugs

```python
drugs_under_18 = average_minutes(at_or_under_18, "120302")
drugs_in_20s = average_minutes(adults_in20s, "120302")
drugs_in_30s = average_minutes(adults_in30s, "120302")
drugs_in_40s = average_minutes(adults_in40s, "120302")
drugs_in_50s = average_minutes(adults_from_50_to_64, "120302")
drugs_above_65 = average_minutes(adults_65_up, "120302")
male_drugs = average_minutes(male, "120302")
female_drugs = average_minutes(female, "120302")
male_vs_female = [male_drugs, female_drugs]
ageing_with_drugs = [drugs_under_18, drugs_in_20s, drugs_in_30s, drugs_in_40s, drugs_in_50s, drugs_above_65]

#plt.plot(ageing_with_drugs, kind='bar')
labels2 = ['Male', 'Female']
labels1 = ['Under 18', '20s', '30s', '40s', '50 - 65', '65+']
male_vs_female = pd.Series(male_vs_female)
ageing_with_drugs = pd.Series(ageing_with_drugs, index=labels1)
ageing_with_drugs.plot(color='black', label = "Average Daily Drug Use")
plt.hlines(male_drugs, xmin=0, xmax=5, color='blue', linestyles = "dashed", label = "Average Male -- 0.39")
plt.hlines(female_drugs, xmin=0, xmax=5, color='red', linestyles = "dashed", label = "Average Female -- 0.21")
plt.ylim(ymin = 0, ymax = .5)
plt.xlabel("Age", fontsize = 15)
plt.ylabel("Minutes per Day", fontsize = 15)
plt.title("Average Time Spent Each Day Using Drugs\n", fontsize = 25)
plt.legend()
plt.show()
```

### Drug use was much lower than expected. The Average American spends less then 1 min a day using drugs/tobacco.

```python
def average_minutes(data, activity_code):
    cols = activity_columns(data, activity_code)
    activity_data = data[cols]
    activity_sums = activity_data.sum(axis=1)
    data = data[['TUFINLWGT']]
    data['minutes'] = activity_sums
    data = data.rename(columns={"TUFINLWGT": "weight"})
    data['weighted_minutes'] = data.weight * data.minutes
    return data.weighted_minutes.sum() / data.weight.sum()
```

```python
def activity_columns(data, activity_code):
    """For the activity code given, return all columns that fall under that activity."""
    col_prefix = "t{}".format(activity_code)
    return [column for column in data.columns if re.match(col_prefix, column)]
```

# Happiness and Health - Socail Factors?

After finding the wellbeing module I decided to take a look at how social interaction affects one's health and wellbeing.

First I looked at groups of people who had spend less than one hour with another person and compared them to a group that had spent less than one hour alone.

```python
lonely_people = respondent[respondent["TRTALONE_WK"] >= 1380]
lonely_people_list = []
for person in range(len(lonely_people)):
    lonely_people_list.append(lonely_people.iget_value(person, 0))
```

```python
lonely_well_being = well_being[well_being["TUCASEID"].isin(lonely_people_list)]
average_health_lonely = lonely_well_being["WEGENHTH"].mean()
ladder_question_lonely = lonely_well_being["WECANTRIL"].mean()
```

```python
popular_people = respondent[respondent["TRTALONE_WK"] <= 60]
popular_people_list = []
for person in range(len(popular_people)):
    popular_people_list.append(popular_people.iget_value(person, 0))
```

```python
popular_well_being = well_being[well_being["TUCASEID"].isin(popular_people_list)]
average_health_popular = popular_well_being["WEGENHTH"].mean()
ladder_question_popular = popular_well_being["WECANTRIL"].mean()
```

```python
social_interaction_data = [average_health_lonely, average_health_popular, ladder_question_lonely, ladder_question_popular]
```

From the chart below we can see that people who spent less than one hour alone per day rated their overall health better than those who spent less than an hour per day with others.  Also, the life satisfaction rating was much higher for those who did not spend much time alone.

```python
plt.xticks(np.arange(4)+.25, ["Avg Health of Persons < 60 Min W/ Others", "Avg Health Persons < 60 Min Alone",
                          "Avg Life Satisfaction < 60 Min W/ Others", "Avg Life Satisfaction < 60 Min Alone"],
           rotation="vertical")
plt.bar(np.arange(4), social_interaction_data, width=.5)
plt.title("Average Ratings of Health and Life Satisfaction Based on Social Interaciton\n"
          "Health - 1 Excellent - 5 Poor   "
          "Satisfaction - 0 Dissatisfied - 10 Extremely Satisfied")
plt.show()
```

```python
health_wellness_time_alone = well_being[["TUCASEID", "WEGENHTH", "WECANTRIL"]].join(respondent["TRTALONE_WK"])
```

```python
hwta_groups = []
hwta_groups.append(health_wellness_time_alone[health_wellness_time_alone["TRTALONE_WK"].isin(range(0, 61))])
for time in range(121, 1441, 60):
    hwta_groups.append(health_wellness_time_alone[health_wellness_time_alone["TRTALONE_WK"].isin(range(time-60, time))])
```

```python
health = []
satisfaction = []
for group in hwta_groups:
    health.append(group["WEGENHTH"].mean())
    satisfaction.append(group["WECANTRIL"].mean())
```

Taking a look at how time spent alone affects health and life satisfaction we can see that the effects are fairly constant until around the 22 hour mark.  After that it seems that there is a significant drop in overall health and satisfaction.

```python
labels = []
plt.title("Health as a Function of Time Spent Alone")
for time in range(61, 1441, 60):
    labels.append("<= {} minutes alone".format(time - 1))
plt.xticks(np.arange(24), labels, rotation="vertical")
plt.plot(health, label="Health, Good<Bad", color="purple")
plt.axis([0, 23, 1, 5])
plt.legend()
plt.show()
```

```python
plt.title("Satisfaction as a Function of Time Spent Alone")
plt.xticks(np.arange(24), labels, rotation="vertical")
plt.plot(satisfaction, label="Life Satisfaction, Bad<Good", color="black")
plt.axis([0, 23, 1, 10])
plt.legend()
plt.show()
```

What other factors play into a persons satisfaction with their life?  I decided to take a look at what activities the most dissatisfied people (Those rating themselves 0) and the most satisfied (Those rating themselves 10) took part in to see if there is a correlation between any of these activities and overall life satisfaction.

```python
most_dissatisfied_sport_rec = average_minutes(most_dissatisfied, "1301")
most_satisfied_sport_rec = average_minutes(most_satisfied, "1301")
most_dissatisfied_social = average_minutes(most_dissatisfied, "12")
most_satisfied_social = average_minutes(most_satisfied, "12")
most_dissatisfied_work = average_minutes(most_dissatisfied, "05")
most_satisfied_work = average_minutes(most_satisfied, "05")
most_dissatisfied_grooming = average_minutes(most_dissatisfied, "0102")
most_satisfied_grooming = average_minutes(most_satisfied, "0102")
```

```python
dissatisfied_data = [most_dissatisfied_sport_rec, most_dissatisfied_social, most_dissatisfied_work,
                     most_dissatisfied_grooming]

satisfied_data = [most_satisfied_sport_rec, most_satisfied_social, most_satisfied_work,
                  most_satisfied_grooming]

fig, ax = plt.subplots()

index = np.arange(4)
bar_width = 0.35

opacity = 0.4

rects1 = plt.bar(index, dissatisfied_data, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Dissatisfied People')

rects2 = plt.bar(index + bar_width, satisfied_data, bar_width,
                 alpha=opacity,
                 color='y',
                 label='Satisfied People')

plt.xlabel('Activity')
plt.ylabel('Average Minutes')
plt.title('A Look at Average Activity Times for Most and Least Satisfied People')
plt.xticks(index + bar_width, ('Playing Sports or Exercising', 'Socializing, Relaxing, Leisure', 
                               'Working', 'Grooming'), rotation="vertical")
plt.legend()

plt.tight_layout()
plt.show()
```

In the above chart we can see that dissatisfied people exercise less, socialize less and work more on average than those who are the most satisfied.  The positive group spends over twice as long exercising each day.

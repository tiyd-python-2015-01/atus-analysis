```python
def activity_columns(data, activity_code):
    """For the activity code given, return all columns that fall under that activity."""
    col_prefix = "t{}".format(activity_code)
    return [column for column in data.columns if re.match(col_prefix, column)]
```

```python
activity_columns(childless_women, "120307")
```

```python
def average_minutes(data, activity_code):
    cols = activity_columns(data, activity_code)
    activity_data = data[cols]
    activity_sums = activity_data.sum(axis=1)
    data = data[["TUFINLWGT"]]
    data["minutes"] = activity_sums
    data = data.rename(columns={"TUFINLWGT": "weight"})
    data['weighted_minutes'] = data.weight * data.minutes
    return data.weighted_minutes.sum() / data.weight.sum()
```

```python
"""The number of minutes per day a woman with no children plays games. Not very much."""

childless_women_game_time = average_minutes(childless_women, "120307")
print (childless_women_game_time)
```

### 1. For how many minutes do adult readers read per day?

```python
adults = summary[summary.TEAGE>=18]
adults.head()
```

```python
average_minutes(adults, "120312") #avg minutes read by adults per day
```

```python
adults_and_reading = adults[["TEAGE", "t120312"]]
adult_readers = np.nonzero(adults_and_reading["t120312"])
adult_readers
```

```python
for i in adult_readers:
    total_read_minutes = sum(i)

total_read_minutes / len(adult_readers[0]) / 60
```

### Imported functions:

```python
def activity_columns(data, activity_code):
    """For the activity code given, return all columns that fall under that activity."""
    col_prefix = "t{}".format(activity_code)
    return [column for column in data.columns if re.match(col_prefix, column)]
```

```python
def average_minutes(data, activity_code):
    """Gets the dataframe and activity code, builds a list of codes using activity_columns.
    Sums the minutes of the activities, weights the minutes and returns a weighted sum. """
    cols = activity_columns(data, activity_code)
    activity_data = data[cols]
    activity_sums = activity_data.sum(axis=1)
    data = data[['TUFINLWGT']]
    data['minutes'] = activity_sums
    data = data.rename(columns={"TUFINLWGT": "weight"})
    data['weighted_minutes'] = data.weight * data.minutes
    try:
        return data.weighted_minutes.sum() / data.weight.sum()
    except:
        return 0
```

```python
def average_counts(data, activity_code):
    def row_count(row_series):
        count = 0
        row_list = list(row_series)
        for col in row_list:
            try:
                if col.ix[0,0] > 0:
                    count += 1
            except:
                pass
        return count
    
    cols = activity_columns(data,activity_code)
    activity_data = data[cols]
    count_list = [row_count(row) for row in activity_data.iterrows()]
    data = data[['TUFINLWGT']]
    data['counts'] = count_list
    data = data.rename(columns={"TUFINLWGT": "weight"})
    data['weighted_count'] = data.weight * data.counts
    try:
        return data.weighted_count.sum() / data.weight.sum()
    except:
        return 0
```

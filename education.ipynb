This analysis will look at how different population groups spend their time in terms of getting additional education. I will be looking specifically at:
- Education to get a degree
- Education for personal interest

### Data Reduction

Columns to use in Analysis:

- TUCASEID:  Identifies each Household
- TULINENO: Identifies each person in the household.  Its value is 1 for the person who answered  the survey


From Roster Dataset:
- TEAGE: Respondent Age
- TESEX: Gender
    - 1 Male
    - 2 Female

From Respondent Dataset:

- TEHRUSL1: How many hours do you work a week?
- TEIO1COW: Individual class of worker (government, private, Self employed,etc)
   Valid Entries: 
   - 1 Government federal
   - 2 Government, state
   - 3 Government, local
   - 4 Private, for profit
   - 5 Private, nonprofit
   - 6 Self-employed, incorporated
   - 7 Self-employed, unincorporated 8 Without pay

From Activities Dataset:
- TRCODE: 6 digit code
- TUACTDUR: Duration of activity in minutes

Activity Codes of interest

- 060101: Taking classes for a degree
- 060102:Taking Classes for personal experience

```python
resp_reduced = respondent[['TUCASEID',
                           'TUFINLWGT',
                           'TEHRUSL1',
                           'TEIO1COW',
                           ]]

for invalid in range (-1,-10,-1):
    resp_reduced.replace(invalid, np.nan,inplace = True)
resp_reduced.head()
```

```python
dem_reduced = demographics[['TUCASEID','TULINENO','TEAGE','TESEX']]
#The person that responds the survey is always TULINENO = 1 but the 
#file has demographics for all people in the household. 

dem_reduced = dem_reduced[dem_reduced['TULINENO'] == 1]
del dem_reduced['TULINENO']
for invalid in range (-1,-10,-1):
    dem_reduced.replace(invalid, np.nan,inplace = True)
dem_reduced.head()
```

```python
codes_desired = [60101, 60102]
act_reduced = activities[activities['TRCODE'].isin(codes_desired)]
act_reduced = act_reduced[['TUCASEID', 'TRCODE', 'TUACTDUR']]
act_reduced.head()
```

### Data Merging

Merge demographics with respondents by TUCASEID

```python
resp_dem = pd.merge(dem_reduced, resp_reduced, on=['TUCASEID'])
resp_dem.sort_index(inplace=True, by='TUCASEID')
resp_dem.head()
```

Merging respondent/demographics with Activities by TUCASEID

```python
all_data = pd.merge(resp_dem, act_reduced, on=['TUCASEID'])
all_data['weighted_minutes'] = all_data['TUFINLWGT'] * all_data['TUACTDUR']
all_data.head()
```

## 2013 Analysis. This part of the report analyzes data for year 2013

```python
plt.rc('figure', figsize=(10,6))
xlabel = "Education Type"
ylabel = "Average Hours"
axis_labels = ['Degree','Personal']
```

```python
def prepare_data(data, key):
    data = data.groupby(['TRCODE', key]).sum()
    data.columns = ['weight_sum', 'weight_min_sum']
    data['weighted_avg'] = (data['weight_min_sum'] / 
                            data['weight_sum']/60).round(decimals=1)
    ax = data.weighted_avg.unstack(level=1).plot(kind='bar')
    return data, ax

def plot_bars(ax,
              legend_labels,
              axis_labels,
              xlabel,
              ylabel,
              title):
    patches, labels = ax.get_legend_handles_labels()
    ax.legend(patches, legend_labels, loc='best')
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.set_xticklabels(axis_labels, rotation=45) 
```

### Year 2013 - Differences in Education by Gender

```python
sum_data=all_data[['TESEX','TUFINLWGT','TRCODE','weighted_minutes']]
sum_data, ax = prepare_data(sum_data,'TESEX')
title = "Average Hours Spent in Education by Gender and Education Type"
legend_labels=['Male','Female']
plot_bars(ax, legend_labels, axis_labels, xlabel, ylabel, title )
```

###### There is no significant different among the genders in terms of the time spent acquiring additional education. In general, people study more to get a degree than for personal development

## Year 2013 - Differences in Education by Employer Type

```python
sum_data=all_data[['TRCODE','TEIO1COW','TUFINLWGT','weighted_minutes']]
sum_data, ax = prepare_data(sum_data,'TEIO1COW')
title = "Average Hours Spent in Education by Employer Type"
legend_labels=['Fed Gov',
               'State Gov',
               'Local Gov',
               'Priv/Profit',
               'Priv/NonProfit',
               'Self/Inc',
               'Self/NotInc',
               'No Pay']
plot_bars(ax, legend_labels, axis_labels, xlabel, ylabel, title )
```

###### In 2013: - Employees in the Federal Government only pursued education for personal development, while employees in State and Local government pursued both types of education - Entrepreneurs who are incorporated pursued education for a degree. Entrepreneurs not incorporated did not pursue any additional education - Employees in the private sector in for profit organizations pursued both types of education vs. those in the nonprofit sector who only pursued education towards a degree

## Year 2013 - Differences in Education by Age

```python
sum_data = all_data[['TRCODE','TEAGE','TUFINLWGT','weighted_minutes']]
sum_data['bin']= pd.cut(sum_data['TEAGE'],[20,35,50,65,80,95],retbins=True)[0]
del sum_data['TEAGE']
sum_data, ax = prepare_data(sum_data,'bin')
title = "Average Hours Spent in Education by Age Group"
legend_labels=['20-34', '35-49', '50-64', '65-79', '80-94']
plot_bars(ax, legend_labels, axis_labels, xlabel, ylabel, title )
```

###### People younger than 65 pursue both education towards a degree as well as for personal development. It is interesting to notice that older people in this group spend as much time as the younger ones studying towards a degree. For the 65+ group, education for personal development is the exclusive priority.

## Year 2013 - Differences in Education by how many hours people work

```python
sum_data = all_data[['TRCODE','TEHRUSL1','TUFINLWGT','weighted_minutes']]
sum_data['bin']= pd.cut(sum_data['TEHRUSL1'],[20,40,60,80,100],retbins=True)[0]
del sum_data['TEHRUSL1']
sum_data, ax = prepare_data(sum_data,'bin')
legend_labels=['20-39', '40-59', '60-79','80-99']
title = "Average Hours Spent in Education by Number of Hours spent at work and by Education Type"
plot_bars(ax, legend_labels, axis_labels, xlabel, ylabel, title )
```

###### In this sample, people who work 40 to 59 hours spend about the same time studying as the people who work less than 40 hours. It is interesting to see people who are working between 80 and 99 hours also finding the time to pursue education towards a degree.

# Trend Analysis - Years 2003 - 2013

## Data Loading

```python
trespondent = pd.read_csv("atusdata/atusresp_0313.dat")
```

```python
tdemographics = pd.read_csv("atusdata/atusrost_0313.dat")
```

```python
tactivities = pd.read_csv("atusdata/atusact_0313.dat")
```

## Data Reduction

```python
tresp_reduced = trespondent[['TUCASEID',
                           'TUFNWGTP',
                           'TEHRUSL1',
                           'TEIO1COW',
                           ]]

for invalid in range (-1,-10,-1):
    tresp_reduced.replace(invalid, np.nan,inplace = True)
tresp_reduced.head()
```

```python
tdem_reduced = tdemographics[['TUCASEID','TULINENO','TEAGE','TESEX']]
#The person that responds the survey is always TULINENO = 1 but the 
#file has demographics for all people in the household. 

tdem_reduced = tdem_reduced[tdem_reduced['TULINENO'] == 1]
del tdem_reduced['TULINENO']
for invalid in range (-1,-10,-1):
    tdem_reduced.replace(invalid, np.nan,inplace = True)
tdem_reduced.head()
```

```python
codes_desired = [60101, 60102]
tact_reduced = tactivities[tactivities['TRCODEP'].isin(codes_desired)]
tact_reduced = tact_reduced[['TUCASEID', 'TRCODEP', 'TUACTDUR']]
tact_reduced.head()
```

## Data Merging

```python
tresp_dem = pd.merge(tdem_reduced, tresp_reduced, on=['TUCASEID'])
tresp_dem.sort_index(inplace=True, by='TUCASEID')
tresp_dem.head()
```

```python
tall_data = pd.merge(tresp_dem, tact_reduced, on=['TUCASEID'])
tall_data['weighted_minutes'] = tall_data['TUFNWGTP'] * tall_data['TUACTDUR']
tall_data['year'] = tall_data['TUCASEID']//10000000000
```

## Exploring how trends have changed during 2003 and 2013

```python
def prepare_multi_year_data(data, key):
    data= tsum_data.groupby(['year','TRCODEP',key]).sum()
    data.columns = ['weight_sum', 'weight_min_sum']
    data['weighted_avg'] = (data['weight_min_sum'] / 
                            data['weight_sum']/60).round(decimals=1)
    del data['weight_min_sum']
    del data['weight_sum']
    data.reset_index(inplace=True)
    data.set_index(['year',key],inplace=True)
    data1 = data [data['TRCODEP'] == 60101]['weighted_avg'].unstack(level=1)
    data2 = data [data['TRCODEP'] == 60102]['weighted_avg'].unstack(level=1)
    return data1, data2
    
 
def prepare_plot_area():
    fig = plt.figure(figsize=(16,8))
    ax1 = fig.add_subplot(211)
    ax2 = fig.add_subplot(212)
    fig.subplots_adjust(hspace=.5)
    return ax1, ax2

    
def plot_lines(data,
               ax,
              legend_labels,
              xlabel,
              ylabel,
              title):
    x_labels= data.index
    ax.set_title(title)
    ax.set_ylabel(ylabel)
    ax.set_xlabel(xlabel)
    ax.set_xticks(np.arange(0,12))
    ax.set_xticklabels(x_labels)
    ax.plot(data)
    patches, labels = ax1.get_legend_handles_labels()
    legend_labels = legend_labels
    ax.legend(legend_labels, loc='best')   
```

## Trends in Education by Gender - Years 2003 -2013

```python
ax1, ax2 = prepare_plot_area()
tsum_data = tall_data[['year','TESEX','TUFNWGTP','TRCODEP','weighted_minutes']]
data1, data2 = prepare_multi_year_data(tsum_data,'TESEX')
legend_labels = ['M','F']

plot_lines(data1,
           ax1,
           legend_labels,
           "Years",
           "Average Hours",
           'Trends in Education towards a degree by Gender. Years 2003 - 2013')
plot_lines(data2,
           ax2,
           legend_labels,
           "Years",
           "Average Hours",
           'Trends in Education for Personal Devleopment by Gender. Years 2003 - 2013')
       
```

###### Time spent acquiring education towards a degree seem to follow similar trends for males and females with males spending on average more time than females in this pursuit.  There is an interesting drop in males pursuing education for personal development in the year 2008 (financial crisis)

## Trends in Education by Employer Type Years 2003 - 2013

```python
ax1, ax2 = prepare_plot_area()
tsum_data = tall_data[['year','TEIO1COW','TUFNWGTP','TRCODEP','weighted_minutes']]
data1, data2 = prepare_multi_year_data(tsum_data,'TEIO1COW')
legend_labels=['Fed Gov', 'State Gov', 'Local Gov', 'Priv/Profit', 'Priv/NonProfit', 'Self/Inc','Self/NotInc','No Pay']

plot_lines(data1,
           ax1,
           legend_labels,
           "Years",
           "Average Hours",
           'Trends in Education towards a degree by Employer Type. Years 2003 - 2013')
plot_lines(data2,
           ax2,
           legend_labels,
           "Years",
           "Average Hours",
           'Trends in Education for Personal Development by Employer Type. Years 2003 - 2013')
       
```

###### Education towards a degree had its highest point with Federal Government employees in 2004. In latest years trends are very similar for all the different employer types except for incorporated enrepreneurs who have shown a marked decrease between 2012 and 2013. For Personal develeopment educaion, it is worth noting how the peak was for government employees at the local level in the year of the financial crisis: 2009.

## Trends by Age - Years 2003 - 2013

```python
ax1, ax2 = prepare_plot_area()
tsum_data = tall_data[['year','TRCODEP','TEAGE','TUFNWGTP','weighted_minutes']]
tsum_data['bin']= pd.cut(tsum_data['TEAGE'],[20,35,50,65,80,95],retbins=True)[0]
del tsum_data['TEAGE']
data1, data2 = prepare_multi_year_data(tsum_data,'bin')
legend_labels=['20-34', '35-49', '50-64', '65-79', '80-94']

plot_lines(data1,
           ax1,
           legend_labels,
           "Years",
           "Average Hours",
           'Trends in Education towards a degree by Age Group. Years 2003 - 2013')
plot_lines(data2,
           ax2,
           legend_labels,
           "Years",
           "Average Hours",
           'Trends in Education for Personal Development by Age Group. Years 2003 - 2013')
```

###### Trends are similar throughout the years. Nothing significant is noticed.

## Trends by Number of Hours Worked - Years 2003 - 2013

```python
ax1, ax2 = prepare_plot_area()
tsum_data = tall_data[['year','TRCODEP','TEHRUSL1','TUFNWGTP','weighted_minutes']]
tsum_data['bin']= pd.cut(tsum_data['TEHRUSL1'],[20,35,50,65,80,95],retbins=True)[0]
tsum_data.tail(50)
del tsum_data['TEHRUSL1']
data1, data2 = prepare_multi_year_data(tsum_data,'bin')
legend_labels=['20-39', '40-59', '60-79','80-99']

plot_lines(data1,
           ax1,
           legend_labels,
           "Years",
           "Average Hours",
           'Trends in Education towards a degree by Hours Worked. Years 2003 - 2013')
plot_lines(data2,
           ax2,
           legend_labels,
           "Years",
           "Average Hours",
           'Trends in Education for Personal Development by Hours Worked. Years 2003 - 2013')
```

###### During the financial crisis people that worked less than 39 hours also allocated a significant amount of time to education towards personal development

```python
summary = pd.read_csv("atusdata/atussum_2013/atussum_2013.dat")
well_being = pd.read_csv("atusdata/wbresp_2013/wbresp_2013.dat")
well_being_activites = pd.read_csv("atusdata/wbact_2013/wbact_2013.dat")
activities = pd.read_csv("atusdata/atusact_2013/atusact_2013.dat")
respondents = pd.read_csv("atusdata/atusresp_2013/atusresp_2013.dat")
cps_data = pd.read_csv("atusdata/atuscps_2013/atuscps_2013.dat")
respondent_crit = (cps_data.TULINENO == 1)
summary_respondents = pd.merge(summary, respondents)
well_being_data = pd.merge(well_being, well_being_activites)
well_being_res = pd.merge(summary_respondents, well_being_data)
well_being_res_act = pd.merge(well_being_res, activities)
well_being_res_act.TRTIER2 = well_being_res_act.TRTIER2.astype(str)
well_being_res_act.info()
cps_data_fixed = cps_data[respondent_crit]
well_being_res_act = pd.merge(well_being_res_act, cps_data_fixed)
```

Pertinent columns:
* TUFINLWGT - statistical weight of respondent
* TRYHHCHILD - age of youngest child in household
* TEAGE - age of respondent
* TESEX - sex of respondent
* TELFS - working status of respondent
* TRCHILDNUM - number of children in household
* TELFS - employment status
* TRTIER2 - activity codes
* WRTELIG - total amount of time in eligible activites
* WUHAPPY - amount of happiness goes from 0 - 6
* WUFNACTWTC - weight used when determining wellbeing

TELFS numbers:
    * 1- Employed - at work
    * 2- Employed - absent
    * 3- Unemployed - on layoff
    * 4- Unemployed - looking
    * 5- Not in labor force
